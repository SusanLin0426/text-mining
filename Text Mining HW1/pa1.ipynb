```python
# Import necessary libraries
import os
import nltk
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Switch to the working directory path
os.chdir('/content/gdrive/My Drive/Colab Notebooks') # Need to set the specified folder path
# os.listdir()
```

```python
# Get the directory where the code file is located
current_directory = os.getcwd()
input_folder_name = 'PA1-data'
output_folder_name = 'output'

# Set the absolute path of the folder
input_folder = os.path.join(current_directory, input_folder_name)
output_folder = os.path.join(current_directory, output_folder_name)


# Create a list to store the content of each document
docs = []

# Read and process each document
for filename in os.listdir(input_folder):
    with open(os.path.join(input_folder, filename), 'r', encoding='utf-8') as file:
        text = file.read()
        docs.append(text)
# Download the stopwords list
nltk.download('stopwords')

from nltk.corpus import stopwords

# Set English stopwords
stop_words = stopwords.words('english')

# Initialize the TF-IDF vectorizer: Lowercase & Filter out English stopwords
tfidf_vectorizer = TfidfVectorizer(lowercase=True, stop_words=stop_words)


# Get the TF-IDF vectors for all documents together
tfidf_matrix = tfidf_vectorizer.fit_transform(docs)

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# Write the TF-IDF vector of each document to separate files
for i, tfidf_vector in enumerate(tfidf_matrix.toarray()):
    output_file = os.path.join(output_folder, f"{i+1}.vec")
    with open(output_file, 'w') as vec_file:
        for value in tfidf_vector:
            vec_file.write(f"{value:.6f}\n")
```

```python
# Calculate the cosine similarity between Document 1 and Document 2
doc1_vector = tfidf_matrix[0]
doc2_vector = tfidf_matrix[1]

# Load the TF-IDF vector of Document 1
document1_file = os.path.join(current_directory, output_folder_name,"1.vec")
with open(document1_file, 'r') as file:
    document1_tfidf = [float(line.strip()) for line in file]

# Load the TF-IDF vector of Document 2
document2_file = os.path.join(current_directory, output_folder_name, "2.vec")
with open(document2_file, 'r') as file:
    document2_tfidf = [float(line.strip()) for line in file]

print(f"TF-IDF vector of documents 1: {document1_tfidf}")
print(f"TF-IDF vector of documents 2: {document2_tfidf}")

print("-------------------------------------------------------------------------------")
# Calculate the cosine similarity between Document 1 and Document 2
cosine_sim = cosine_similarity(doc1_vector, doc2_vector)
print(f"Cosine Similarity between Document 1 and Document 2 is {cosine_sim[0][0]:.4f}")
```
